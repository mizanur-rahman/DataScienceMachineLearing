##### NESTED FUNCTION

```python
#nested function
def square():
    """ return square of value """
    def add():
        """ add two local variable """
        x = 2
        y = 3
        z = x + y
        return z
    return add()**2
print(square())
```

def f(*args):
  *args can be one or more
  def f(** kwargs)
  **kwargs is a dictionary"
  

```python
# default arguments
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))
# what if we want to change default arguments
print(f(5,4,3))
```

```python
# flexible arguments *args
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)
# flexible arguments **kwargs that is dictionary
def f(**kwargs):
    """ print key and value of dictionary"""
    for key, value in kwargs.items():               # If you do not understand this part turn for loop part and look at dictionary in for loop
        print(key, " ", value)
f(country = 'spain', capital = 'madrid', population = 123456)
```

##### RESAMPLING PANDAS TIME SERIES
1. Resampling: statistical method over different time intervals
2. Needs string to specify frequency like "M" = month or "A" = year
3. Downsampling: reduce date time rows to slower frequency like from daily to weekly
4. Upsampling: increase date time rows to faster frequency like from daily to hourly
5. Interpolate: Interpolate values according to different methods like ‘linear’, ‘time’ or index’
6. https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.interpolate.html

```python
# We will use data2 that we create at previous part
data2.resample("A").mean()
```

```python
# Lets resample with month
data2.resample("M").mean()
# As you can see there are a lot of nan because data2 does not include all months
```
